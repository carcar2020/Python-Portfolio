## Documentation

**Purpose:** This notebook visualizes data for quantitative and categorical Googleplaystore Apps data. It also has summarized data for a grouped dataframe and joined dataframe.



This notebook focuses on data regarding user reviews and sentiment including: 'translated review(comment)', Sentiment, Sentiment polarity, and Sentiment subjectivity for file `googleplaystore_user_reviews.csv`, although only data regarding Sentiment and App appearance in user review is analyzed( number of times the App is present in `googleplaystore_user_reviews.csv`.
```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('googleplaystore_user_reviews.csv')
df2 = pd.read_csv('googleplaystore.csv')
```

```python
# Quantitative Data based on total reviews for the top 10 most reviewed Apps on file user_reviews
total_App_Rev = df1['App'].value_counts()
#total_App_Rev['10 Best Foods for You']
total_App_Rev = total_App_Rev.sort_values(ascending = False).head(10)
print(total_App_Rev)


total_App_Rev = df1['App'].value_counts()
total_App_Rev = total_App_Rev.sort_values(ascending = False).head(10)
plt.barh(total_App_Rev.index, total_App_Rev.values)
plt.title('Top 10 Most Reviewed Apps')
plt.xlabel("Reviews")

#total_App_Rev['10 Best Foods for You']
```

## Quantitative Data ##
**Top 10 Most Reviewed Apps :**
The above horizontal barchart was organized based on that number of Reviews a certain app got in order of lowest reviewed App on top and highest reviewed on bottom. We can see that user decided to leave reviews mainly for CBS Sports App, Bowmasters, and Angry Birds.

```python
# Categorical Data based on total sentiment for Positive and Negative 

num1 = df1[df1['Sentiment'] == 'Positive']['Sentiment'].value_counts()
num2 = df1[df1['Sentiment'] == 'Negative']['Sentiment'].value_counts()
x = [23998, 8271]
neww = pd.DataFrame(x)

plt.bar(neww.index, x)
plt.xlabel('Positive           Negative ')
plt.ylabel('Sentiment Total')
plt.title(' Positive Vs Negative ')
```

### Categorical Data Summary

The above barchart was composed from data regarding the total value counts for where Sentiment was Negative and Positive for user_reviews and we can see that the overall sentiment for this Googleplaystore Reviews dataset is almost 3x more positive when compare to negative sentiment on this data set.


```python
# Grouped Data Summary 

nn = df2.groupby('Category').mean()
nnw = nn[ nn['Rating'] <= 5].sort_values(by = 'Rating', ascending = False).head(10)
print(nnw.head(20))

nnw.plot.pie(subplots =True, figsize = (10, 10))
plt.title('Highest Average Rating: Based on Category')
```

### Grouped Data Summary

The data above summarizes the mean for the ratings based on category. The piechart was visualized based on the mean values we got after grouping the ratings by categories. This data is displayed from highest mean rating value to lowest. With the piechart we can only build a conclusion that the mean rating for each categorical is quite similar in rating.

```python
t1 = df1['App']
t2 = df2['App']
t1_t2 = pd.merge(t1, t2, how = 'inner', on= 'App')
print(t1_t2.drop_duplicates())
```

### Joined Tables Data Summary

The data listed above are all the Applications in the dataset of googleplaystore that were reviewed in any way. The use of drop_duplicates() was used to only allow for apps to appear only once since in one csv file some apps are reviewed multiple times and appear more then once.
